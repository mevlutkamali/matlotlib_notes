**Grafik Düzenleme**

```python
x = np.arange(0, 10)
y = x ** 2 - 10 * x + 25
```

```python
x, y
```

```python
plt.plot(x, y)
```

.figsize = Sayfanın genişlik ve yüksekligini belirtiyor .

.dpi = Sayfanın görüntü kalitesini arttırıyor .

```python
plt.figure(figsize=(12, 6), dpi=200)

plt.plot(x, y)

plt.title("Grafik Başlıgı", size=15)

plt.xlabel("x", size=10)
plt.ylabel("y", size=10)

plt.xticks(np.arange(min(x), max(x)+1, 1))
plt.yticks(np.arange(min(y), max(y)+1, 2.5))
```

**Limit Alanı Belirleme**

**Legend**

**.legend()** = Bir grafikte çizilen veri serilerini açıklamak için kullanılır . Grafiklerde birden fazla veri serisi (örneğin, çizgi grafiği, çubuk grafiği veya dağılım grafiği) olduğunda, bu serilerin hangi renklere veya sembollere karşılık geldiğini belirtmek amacıyla kullanılır . 

```python
x = np.arange(0, 10)

plt.figure(figsize=(10, 8), dpi=140)

plt.plot(x, 2 * x + x, label="2x + x")
plt.plot(x, x ** 2 - x, label="x ^ 2 - x")

plt.legend(loc=0)

plt.show()
```

loc=

0 -> En iyi yeri seçer .

1 -> Sağ üst .

2 -> Sol üst .

3 -> Sol alt .

4 -> Sağ alt .

 **Çizği Şeklini Değiştirmek**

```python
plt.figure(figsize=(10, 6), dpi=200)

plt.plot(x, x + 1, color="blue", linestyle="-", lw=3)
```

```python
plt.figure(figsize=(10, 6), dpi=200)

plt.plot(x, x + 1, color="black", marker="s", markersize=14)
plt.plot(x, x + 2, color="red", marker="o", markersize=14)
plt.plot(x, x + 3, color="green", marker="1", markersize=14)
plt.plot(x, x + 4, color="blue", marker="+", markersize=14)
```

**Subplot**

**Subplot** işlevi, birden fazla grafik veya alt resmi aynı figür içinde düzenlemek için kullanılır.

```python
fig, axes = plt.subplots(nrows=2, ncols=2)

axes.shape

x = np.arange(0, 10)
y = x + 9 

fig, axes = plt.subplots(1,2)

axes[0].plot(x, y)
axes[1].plot(x, y)
```
